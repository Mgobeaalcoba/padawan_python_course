## 1. Graficos de barras
Se utiliza para comparar valores discretos o categ√≥ricos entre diferentes grupos.

```python
# @title Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# @title Create DataFrame step
df = pd.read_csv("https://raw.githubusercontent.com/Statology/Python-Guides/main/default.csv")
```

```python
# @title Inspection step
df.sample(5)
```

```python
# @title Set auxiliaries variables
x = 'income'
y = 'balance'
category = 'student'
category2 = 'default'
```

```python
# @title Barplot con suma

## Seteo el grafico de barras de suma con apertura de segmento
sns.barplot(data=df,
            x=category,
            y=y,
            estimator='sum')

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel('Sum '+ y)
plt.title('Grafico de barra')

## Muestro el gr√°fico resultante
plt.show()
```

#### La suma de balance no ve con claridad porque est√° escrito de forma exponencial dado que es un n√∫mero muy grande. Podemos recurrir entonces a transformar/agregar nuestro "balance" a/en miles (K) y volver a graficar con el

```python
df["balance_K"] = df["balance"] / 1000
df.sample(5)
```

```python
# @title Repito mi Barplot en miles

## Seteo el grafico de barras de suma
sns.barplot(data=df,
            x=category,
            y=df["balance_K"], # Cambio mi eje Y por la nueva columna creada
            estimator='sum')

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel('Sum balance K')
plt.title('Grafico de barra')

## Muestro el gr√°fico resultante
plt.show()
```

```python
# @title Barplot con promedio

## Grafico de barras de promedio
sns.barplot(data=df,
            x=category,
            y=y,
            estimator='mean')

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel('Average '+y)
plt.title('Grafico de barra')

## Muestro el gr√°fico resultante
plt.show()
```

```python
# @title Barplot de promedio aperturado por segmento

## Grafico de barras de promedio con apertura de segmento
sns.barplot(data=df,
            x=category,
            y=y,
            hue=category2,
            estimator='mean')

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel('Average '+y)
plt.title('Grafico de barra')

## Muestro el gr√°fico resultante
plt.show()
```

## 2. Graficos de Lineas
Se utiliza para mostrar cambios en una variable a lo largo del tiempo o en funci√≥n de otra variable.

```python
# @title Create DataFrame step
df2 = pd.read_csv('https://www.dropbox.com/s/4jgheggd1dak5pw/data_visualization.csv?raw=1', index_col=0)
df2['args'] = np.random.choice([0, 1],size=len(df2.index)) # Creamos una columna nueva con n√∫meros enteros entre 0 y 1 escogidos aleatoriamente.
```

```python
# @title Inspection step
df2.head()
```

```python
# @title Set auxiliaries variables for regression
x = 'height'
y = 'slice'
category = 'args'
```

```python
# @title Ordeno mi dataframe de forma ascendente(por default) por mi columna 'height
df2 = df2.sort_values(x)
df2.head()
```

```python
# @title Lineplot

## Grafico de linea para poder identificar distribuci√≥n
sns.lineplot(data=df2,
             x=x,
             y=y,
             linestyle='--',
             linewidth=2)

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)
plt.title('Gr√°fico de lineas')

## Muestro el gr√°fico resultante
plt.show()
```

```python
# @title Lineplot aperturado por "args"

## Grafico de linea agregando categoria
sns.lineplot(data=df2,
             x=x,
             y=y,
             linestyle='--',
             linewidth=2,
             hue=category)

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)
plt.title('Gr√°fico de lineas')

## Muestro el gr√°fico resultante
plt.show()
```

## Grafico de Dispersion
Se utiliza para mostrar la relaci√≥n entre dos variables num√©ricas

```python
# @title Create DataFrame step
df3 = pd.read_csv("https://raw.githubusercontent.com/Statology/Python-Guides/main/default.csv")
df3.tail(3)
```

```python
# @title Set auxiliaries variables
y = 'income'
x = 'balance'
category = 'student'
categroy2= 'default'
```

```python
# @title Lmplot

## Grafico de dispersion
sns.lmplot(data=df3,
           x=x,
           y=y,
           scatter_kws={"s": 1})

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)
plt.title('Grafico de dispersion')

## Muestro el gr√°fico resultante
plt.show()
```

```python
# @title Lmplot open by category

## Grafico de dispersion aperturada por categoria
sns.lmplot(data=df3,
           x=x,
           y=y,
           hue=category,
           scatter_kws={"s": 1})

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)
plt.title('Grafico de dispersion')

## Muestro el gr√°fico resultante
plt.show()
```

```python
# @title Lmplot of two columns

## Grafico de dispersion aperturado por categoria seg√∫n el color (hue) y por columna seg√∫n segunda categoria (col)
sns.lmplot(data=df3, x=x, y=y, hue=category, col=categroy2, scatter_kws={"s": 10})

## Seteo los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)

## Muestro el gr√°fico resultante
plt.show()
```

## Grafico de Histograma
Se utiliza para mostrar la distribuci√≥n de una variable num√©rica.

```python
# @title Create DataFrame step
df4 = pd.read_csv("https://raw.githubusercontent.com/Statology/Python-Guides/main/default.csv")
df4.sample(6)
```

```python
# @title Set auxiliaries variables
y = 'income'
x = 'balance'
category = 'student'
```

```python
# @title Histplot

## Histograma por cantidad de bins
sns.histplot(data=df4,
             x=x,
             bins=20,
             color='purple')

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)
plt.title('Histograma')

## Muestro el gr√°fico resultante
plt.show()
```

```python
# @title Histplot open by category

## Agrego apertura por categoria
sns.histplot(data=df, x=x, bins=20, hue=category)

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)
plt.title('Histograma')

## Muestro el gr√°fico resultante
plt.show()
```

```python
# @title Histplot with trends lines

## Agrego lineas de tendencia
sns.histplot(data=df, x=x, bins=20, hue=category, kde=True, multiple='dodge', stat='percent')

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)
plt.title('Histograma')

## Muestro el gr√°fico resultante
plt.show()
```

## Grafico de Boxplot
Se utiliza para mostrar la distribuci√≥n de una variable num√©rica y destacar los valores at√≠picos.

```python
# @title Create DataFrame step
df5 = pd.read_csv("https://raw.githubusercontent.com/Statology/Python-Guides/main/default.csv")
df5.head(7)
```

```python
# @title Set auxiliaries variables
y = 'income'
x = 'balance'
category = 'student'
```

```python
# @title Boxplot open by category

## Construyo mi boxplot
sns.boxplot(data=df5,
            x=category,
            y=y)

## Seteo el titulo y los nombres de los ejes
plt.xlabel(x)
plt.ylabel(y)
plt.title('Boxplot')

## Muestro el gr√°fico resultante
plt.show()
```
